# re.match
```re.match(pattern,string,flags=0)```尝试从字符串的起始位置匹配一个模式，如果不是起始位置匹配成功的话，返回None

## 常规匹配

````python
import re
content = 'Hello 123 4567 World_This is a Regex Demo'
print(len(content))
result = re.match('^Hello\s\d\d\d\s\d{4}\s\w{10}.*Demo$', content)
print(result)
print(result.group())
print(result.span())
````

## 泛匹配

````python
import re
content = 'Hello 123 4567 World_This is a Regex Demo'
result = re.match('^Hello.*Demo$', content)
print(result.group())
````

## 匹配目标

````python
import re
content = 'Hello 1234567 World_This is a Regex Demo'
result = re.match('^Hello\s(\d+)\sWorld.*Demo$', content)
print(result)
print(result.group())
print(result.group(1))#第一个小括号括起来的内容就是group(1)
print(result.span())
````

## 贪婪匹配

````python
import re
content = 'Hello 1234567 World_This is a Regex Demo'
result = re.match('^He.*(\d+).*Demo$', content)
print(result)
print(result.group(1))#只匹配出了7，因为前面的123456被.*匹配了
````

## 非贪婪匹配

````python
import re
content = 'Hello 1234567 World_This is a Regex Demo'
result = re.match('^He.*?(\d+).*Demo$', content)
print(result)
print(result.group(1))
````

## 匹配模式

````python
import re
content = '''Hello 1234567 World_This 
is a Regex Demo'''
result = re.match('^He.*?(\d+).*Demo$', content,re.S)# .不能匹配换位符，所以加re.S
print(result)
print(result.group(1))
````

## 转义

````python
import re
content = 'price is $5.00'
result=re.match('price is \$5\.00',content)# $和.需要加\转义
print(result)
````

**总结：尽量使用泛匹配，使用括号得到匹配目标，尽量使用非贪婪模式，有换行符就用re.S**

# re.search
re.search扫描整个字符串并返回第一个成功的匹配

````python
import re
content = 'Extra stings Hello 1234567 World_This is a Regex Demo'
result = re.search('Hello.*?(\d+).*?Demo', content)
print(result)
print(result.group(1))
````

**总结：尽量使用re.search,因为match要输入开头**

# re.findall
搜索字符串，以列表形式返回全部能匹配的子字符串

````python
import re

html = '''<div id="songs-list">
    <h2 class="title">经典老歌</h2>
    <p class="introduction">
        经典老歌列表
    </p>
    <ul id="list" class="list-group">
        <li data-view="2">一路上有你</li>
        <li data-view="7">
            <a href="/2.mp3" singer="任贤齐">沧海一声笑</a>
        </li>
        <li data-view="4" class="active">
            <a href="/3.mp3" singer="齐秦">往事随风</a>
        </li>
        <li data-view="6"><a href="/4.mp3" singer="beyond">光辉岁月</a></li>
        <li data-view="5"><a href="/5.mp3" singer="陈慧琳">记事本</a></li>
        <li data-view="5">
            <a href="/6.mp3" singer="邓丽君"><i class="fa fa-user"></i>但愿人长久</a>
        </li>
    </ul>
</div>'''

results = re.findall('<li.*?href="(.*?)".*?singer="(.*?)">(.*?)</a>', html, re.S)
print(results)
print(type(results))
for result in results:
    print(result)
    print(result[0], result[1], result[2])
````

````python
import re

results = re.findall('<li.*?>\s*?(<a.*?>)?(\w+)(</a>)?\s*?</li>', html, re.S)
for result in results:
    print(result[1])
````

# re.sub
替换字符串中每一个匹配的子串后的字符串

````python
import re
content = 'Hello 1234567 World_This is a Regex Demo'
content=re.sub('\d+','',content)#将数字替换为空
print(content)
````

````python
#替换后的内容包含原来的内容， 
import re
content = 'Hello 1234567 World_This is a Regex Demo'
content=re.sub('(\d+)',r'\1 8910',content)# \1 表示把第一个括号的内容拿过来
print(content)
````
